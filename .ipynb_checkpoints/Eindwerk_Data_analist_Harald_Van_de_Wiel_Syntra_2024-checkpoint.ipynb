## Bitcoin ETF Data

*We beginnen met de BTC ETF data eerst. We willen graag weten welke tickers we nodig hebben. Dankzij twitter
hebben we een site gevonden waar we alle tickers terug vinden die we nodig hebben*

```python
#Web pagina inladen met request
"""we beginnen met de BTC ETF data eerst. We willen graag weten welke tickers we nodig hebben. Dankzij twitter
hebben we een site gevonden waar we alle tickers terug vinden die we nodig hebben"""

web = requests.get("https://farside.co.uk/?p=997")

#Soup gebruiken om de website in te laden en zo de data uit de pagina info te kunnen halen

soup = BeautifulSoup(web.text, 'html.parser')
```

```python
#Pagina inladen om te kijken wat we nodig hebben
soup.find_all
```

```python
tickerlist = []
etfnamen = soup.find_all('span', class_='tabletext')

# Definieer een lijst van waarden die je niet wilt toevoegen
exclude = ['Date']  # Je kunt deze lijst uitbreiden met andere waarden die je wilt uitsluiten

# Loop door de gevonden <span> elementen en haal de etf ticker naam op
for etf in etfnamen:
    ticker = etf.get_text()
    # Controleer of de huidige waarde niet in de exclude lijst staat voordat je het toevoegt
    if ticker not in exclude:
        tickerlist.append(ticker)  # Voeg ticker toe aan de lijst als het niet uitgesloten is
    if ticker == 'GBTC':  # Controleer of de huidige ticker het laatste gewenste ticker symbool is
        break  # Stop de loop na het toevoegen van de laatste gewenste ticker

# Nu bevat 'tickerlist' alleen de gewenste tickers tot en met 'GBTC', zonder 'Date'
print(tickerlist)
```

## Gold ETF's Data

*Nu gaan we hetzelfde doen voor onze Gold ETF data via de https://etfdb.com/etfs/commodity/gold/ website.*
*Hier gaan we ook lijst maken voor onze Gold ETF tickers en deze opslaan in een lege lijst om daarna met yahoo finance op te roepen.*
*Op de site hebben we reeds gekeken dat DBZ de laatste ticker in de tabel daar is voor onze volledige Gold etf lijst*

```python
goldweb = requests.get("https://etfdb.com/etfs/commodity/gold/")

#Soup gebruiken om de website in te laden en zo de data uit de pagina info te kunnen halen

goldsoup = BeautifulSoup(goldweb.text, 'html.parser')
```

```python
tr_elements = goldsoup.find_all('tr')


# lege lijst goud tickers
goldticker = []

# Loop door de html lijst en voeg de tickers to aan een lege lijst en stop bij 'Click Here to Join to ETF Database Pro for 14 Days Free, Export This Data & So Much More'
for tr in tr_elements:
    a_tag = tr.find('a')
    if a_tag:
        ticker = a_tag.text
        if ticker == 'Click Here to Join to ETF Database Pro for 14 Days Free, Export This Data & So Much More':
            break  
        goldticker.append(ticker) 


print(goldticker)

```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Veronderstel dat 'total_volume' je DataFrame is met datums, GoldVolume, BitcoinVolume, en VolumeVerschil
# Eerst, zorg ervoor dat 'Date' als een datetime object is en ingesteld als index
total_volume['Date'] = pd.to_datetime(total_volume.index)
total_volume.set_index('Date', inplace=True)

# Maak de plot
plt.figure(figsize=(10, 6))  # Stel de grootte van je figuur in

# Teken de lijnen
plt.plot(total_volume.index, total_volume['GoldVolume'], label='Goud Volume', color='gold', linewidth=2)
plt.plot(total_volume.index, total_volume['BitcoinVolume'], label='Bitcoin Volume', color='orange', linewidth=2)
plt.plot(total_volume.index, total_volume['VolumeVerschil'], label='Volume Verschil', color='green', linestyle='--', linewidth=2)

# Voeg titel en labels toe
plt.title('Volume Verschil tussen Goud en Bitcoin', fontsize=16)
plt.xlabel('Datum', fontsize=14)
plt.ylabel('Volume', fontsize=14)

# Draai de datums op de x-as voor betere leesbaarheid
plt.xticks(rotation=45)

# Voeg een legenda toe
plt.legend()

# Voeg een grid toe voor betere leesbaarheid
plt.grid(True, linestyle='--', alpha=0.5)

# Toon de grafiek
plt.tight_layout()  # Past de subplot parameters aan zodat de subplot(s) in het figuur gebied passen
plt.show()
```
